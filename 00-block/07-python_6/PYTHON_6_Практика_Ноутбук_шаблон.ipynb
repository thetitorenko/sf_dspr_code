### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом A и клиентом B на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу «Некорректный ввод».

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
[42, 52, 14]

```

**Пример №2:**
```
Ввод:
108, 138, 42, 52, 14
189, 202, 249, 24, 15, 3

Вывод:
[]

```

**Пример №3:**
```
Ввод:
108, 138, 42, 52, четырнадцать
189, 202, 249, 24, 15, 3

Вывод:
Некорректный ввод

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**4 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**2 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
# Функция для нахождения пересечение двух списков без дубликатов
def pure_intersection(input1, input2):

    # Функция для перевода ввода в список целых чисел
    def get_int_list(input):

        # Разбиваем ввод по запятой и преобразуем в целые числа
        try:
            input_splitted = input.split(', ')
            # Применяем функцию перевода в целое число к каждому элементу списка
            # с помощью map() и переводим результат обратно в список
            input_result = list(map(lambda x: int(x), input_splitted))
            return input_result
        
        # Если ввод не является числом, возвращаем None
        except ValueError:
            return None

    # Переводим ввод в списки
    input1_list = get_int_list(input1)
    input2_list = get_int_list(input2)

    # Проверяем, что оба списка успешно созданы
    if input1_list is None or input2_list is None:
        print('Некорректный ввод')
        return

    # Находим пересечение двух списков с удалением из них дубликатов
    # путем перевода их в множества
    result = set(input1_list).intersection(set(input2_list))

    # Возвращаем результат в виде списка
    return list(result)
    

# Примеры использования функции
print(pure_intersection(input1='108, 138, 42, 52, 14', input2='109, 13, 52, 32, 42, 14, 109'))
print(pure_intersection(input1='108, 138, 42, 52, 14', input2='189, 202, 249, 24, 15, 3'))
print(pure_intersection(input1='108, 138, 42, 52, четырнадцать', input2='189, 202, 249, 24, 15, 3'))
````

## Задание №2

#### Найти минимум и максимум

**Компания:** «Авито»

___

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию `input()`).

Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число словами (например, не «5», а «пять») — в этом случае слово нужно игнорировать.

**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**

```
Ввод:
-2,56 25,002 восемь 19,13 13 -37,5

Вывод:
Minimum: -37.5
Maximum: 25.002
```



**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

**Критерии оценивания:**

- Задача решена корректно, учтены все условия (**4 балла**):
    - программа корректно обрабатывает ввод пользователя и работает на любых входных данных;
    - программа корректно обрабатывает присутствие слов во вводе пользователя;
    - программа корректно рассчитывает минимум и максимум числовой последовательности;
    - решение сопровождено комментариями.
- Задача решена корректно, но не выполнено одно из условий (**2 балл**):
    - программа некорректно обрабатывает ввод пользователя и работает только на конкретном примере входных данных;
    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
# Функция для нахождения минимального и максимального числа в строке
def find_min_max(s):

    # Разбиваем строку по знаку пробела
    s_splitted = s.split()

    # Фильтруем строку убирая все элементы, введенные словами
    s_filtered = list(filter(lambda x: not x.isalpha(), s_splitted))
    
    # Каждый элемент строки переводим в вещественное число,
    # предварительно заменив запятую на точку для избежания ошибки
    s_result = list(map(lambda x: float(x.replace(',', '.')), s_filtered))

    # Выводим результат
    print(f'Minimum: {min(s_result)}')
    print(f'Maximum: {max(s_result)}')


# Примеры использования функции
find_min_max(s='-2,56 25,002 восемь 19,13 13 -37,5')
find_min_max(s='5,5 6,6 семь -10 3,14 -2,72')
find_min_max(s='100 200,5 триста -1000 -500,25')
````

## Задание №3

#### Найти медиану

**Компания:** «Авито»

**Уровень:** Лёгкий

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**
```
Ввод:
1, 5, 2, 3, 6

Вывод:
Median: 3.0
```
```
Ввод:
100, 5, 2, 4, 3, 6

Вывод:
Median: 4.5
```
```
Ввод:

Вывод:
Некорректный ввод
```
```
Ввод:
десять, 10, пять, 7, семь

Вывод:
Некорректный ввод
```


**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**6 балла**):
    - ввод пользователя конвертируется в список из чисел;
    - функция для расчёта медианы работает корректно для списка чётной и нечётной длины;
    - в случае некорректного ввода функция возвращает `None`;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**3 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для расчёта медианы некорректно работает для списка чётной или нечётной длины;
    - в случае некорректного ввода функция отрабатывает с ошибкой;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм расчёта медианы для обоих случаев, результат работы программы зависит от вводимой последовательности чисел и т. д.

````python
# Функция для нахождения медианы в вводе
def find_median(input):

    # Если ввод пустой, выводим сообщение об ошибке
    if not input:
        print('Некорректный ввод')
        return

    # Переводим ввод (строка) в список. Если в вводе есть
    # цифры введенные буквами - выводим сообщение об ошибке
    input_list = []
    for num in input.split(', '):
        if num.isalpha():
            print('Некорректный ввод')
            return
        else:
            input_list.append(float(num))
    
    # Сортируем список по возрастанию
    input_list.sort()

    # Если длина списка нечетная, медиана - это элемент посередине
    if len(input_list) % 2 != 0:
        # Индекс элемента посередине списка - это длина списка деленная на 2
        median = input_list[int(len(input_list) / 2)]
        return median
    
    # Если длина списка четная, медиана - это среднее арифметическое 
    # двух элементов посередине
    else:
        # Индексы двух элементов посередине списка - это длина списка деленная на 2
        half_length = int(len(input_list) / 2)
        # Делаем среза списка с двумя элементами посередине
        middle_nums = input_list[half_length-1:half_length+1]
        # Считаем среднее арифметическое двух элементов
        median = sum(middle_nums) / 2
        return median
    

# Примеры использования функции
print(f'Median: {find_median(input='1, 5, 2, 3, 6')}')
print(f'Median: {find_median(input='100, 5, 2, 4, 3, 6')}')
print(f'Median: {find_median(input='')}')
print(f'Median: {find_median(input='десять, 10, пять, 7, семь')}')
````

При решении задачи вам может пригодиться следующий словарь:

number_word_dict = {

**Подсказка №2:**

При решении задачи вам может пригодиться метод строк `s.startswith(word)`, который определяет, начинается ли строка `s` со строки `word`. Подробнее о нём вы можете узнать в [документации](https://docs-python.ru/tutorial/operatsii-tekstovymi-strokami-str-python/metod-str-startswith/).

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**6 балла**):
    - программа корректно преобразует заданное словами число в число;

     решение сопровождено комментариями.

    - решение не сопровождено комментариями.

- Задача не решена (**0 баллов**).

````python
# Функция для преобразования строки в число
def transform_string_to_integer(s):
    
    # Словарь соответствия начала числительных и их значения
    number_word_dict = {
        "ты": 1_000, "м": 1_000_000,
        "сто": 100, "двес": 200, "трис": 300, "четырес": 400, "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
        "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16, "семн": 17, "восемн": 18, "девятн": 19,
        "двад": 20, "трид": 30, "сор": 40, "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90,
        "дес": 10, "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9, }

    s_splitted = s.split()      # Разбиваем строку на слова-цифры
    result = 0                  # Итоговая сумма
    temp_result = 0             # Промежуточная сумма для чисел до 1_000 или 1_000_000

    # Перебираем слова-цифры в строке
    for num in s_splitted:
        matched = False             # Флаг, чтобы отслеживать совпадение ключа

        # Перебираем ключи словаря с числительными и их значениями
        for key in number_word_dict:

            # Если начало слова-цифры совпадает с ключом словаря,
            if num.startswith(key):
                matched = True          # Меняем значение флага, т.к. есть совпадение

                # Умножаем временную переменную на 1_000 или 1_000_000
                if (number_word_dict[key] == 1_000) or (number_word_dict[key] == 1_000_000):

                    # Если слова-цифра "тысяча" или "миллион" идет без числа перед ним
                    if temp_result == 0:
                        temp_result = 1

                    temp_result *= number_word_dict[key]
                    result += temp_result
                    temp_result = 0     # Сбрасываем промежуточный результат

                # Прибавляем к накопительной переменной значению по ключу
                else:
                    temp_result += number_word_dict[key]
                
                break           # Прекращаем проверку после первого совпадения
        
        # Если не было совпадения, выводим сообщение об ошибке
        if not matched:
            raise ValueError(f'Неизвестное числительное: {num}')
    
    
    result += temp_result       # Добавляем остаток после последнего слова-цифры
    return result               # Возвращаем результат


# Примеры использования функции
print(transform_string_to_integer('один'))
print(transform_string_to_integer('двадцать'))
print(transform_string_to_integer('двести сорок шесть'))
print(transform_string_to_integer('семьсот восемьдесят три тысячи девятьсот девятнадцать'))
print(transform_string_to_integer('шестьсот сорок тысяч семьсот тридцать пять'))
print(transform_string_to_integer('миллион двести тридцать четыре тысячи пятьсот шестьдесят семь'))
````

### Задание №5 (бонусная, не оценивается)

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
# Функция разложения факториала на простые множители
def decompose_factorial(n):
    
    # Функция для нахождения факториала числа
    def factorial(n):
        if n == 0 or n == 1: return 1
        return n * factorial(n-1)
    

    # Функция для проверки простоты числа
    def is_primal(num):
        div_count = 0       # Счетчик делителей
        
        if 2 <= num <= 3:   # Если число 2 или 3,
            return True     # оно простое

        # Проверяем делимость числа на все числа от 1 до исходного
        for i in range(1, num+1):
            if num % i == 0:        # Если число делится на i без остатка,
                div_count += 1      # увеличиваем счетчик делителей
        
        # Если число делится только на себя и на 1,
        if div_count == 2: return True  # оно простое.
        else: return False              # Если неn - не простое


    # Функция для нахождения степени простого множителя в факториале числа n
    # по формуле: floor(n/p) + floor(n/p^2) + floor(n/p^3) + ...
    # пока p^k <= n, где p - простое число, n - факториал числа n
    def get_prime_factors_degree(p, n):
        degree = 0
        k = 1
        while p**k <= n:
            degree += (n // p**k)
            k += 1
        return degree


    # Вычисляем факториал числа n 
    # с проверкой, что n - целое неотрицательное число
    if isinstance(n, int):
        n_factorial = factorial(n)
    else:
        print('Введите целое не отрицательное число')
        return      # Выходим из главной функции

    # Создаем список простых множителей
    prime_nums_list = []
    for i in range(1, n+1):                 # Проверяем все числа от 1 до n
        if is_primal(i):                    # Если число простое,
            prime_nums_list.append(i)       # добавляем его в список

    prime_factors_degree_dict = {}          # Словарь для хранения простых множителей (ключ) и их степеней (знач.)
    for prime_num in prime_nums_list:
        prime_factors_degree_dict[prime_num] = get_prime_factors_degree(prime_num, n)

    prime_factors_str_temp = ''             # Временная строка для хранения простых множителей и их степеней

    # Формируем временную строку с простыми множителями и их степенями,
    # записанными через знак ^, и разделенными пробелами
    for prime_key, degree_value in prime_factors_degree_dict.items():
        # Если степень больше 1, добавляем знак ^ и степень
        if degree_value > 1:
            prime_factors_str_temp += str(prime_key) + '^' + str(degree_value) + ' '
        # Если степень равна 1, просто добавляем число
        else:
            prime_factors_str_temp += str(prime_key) + ' '

    # Формируем готовую строку простых чисел и их степеней,
    # с помощью разделения временной строки на список по пробелам,
    # и соединению его вновь в строку со знаком умножения в качестве разделителя
    prime_factors_str_done = ' * '.join(prime_factors_str_temp.split())

    # Вывод функции
    print(f'{n}! = {n_factorial} = {prime_factors_str_done}')


# Примеры использования функции (с добавлением значения факториала)
decompose_factorial(5)      # 5! = 2^3 * 3 * 5
decompose_factorial(22)     # 22! = 2^19 * 3^9 * 5^4 * 7^3 * 11^2 * 13 * 17 * 19
decompose_factorial(25)     # 25! = 2^22 * 3^10 * 5^6 * 7^3 * 11^2 * 13 * 17 * 19 * 23
````

### Задание №6 (бонусная, не оценивается)

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:
zzzab

step 1: zaba
step 2: zbb
step 3: zc

output:
cz
```

```
input:
hhakafh

step 1: haakfi
step 2: hkfib

output:
hkfib
```

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)
````

````python
import string


# Рекурсивная функция замены дублирующихся букв
def replace_duplicates(s, step=1):
    alphabet = string.ascii_lowercase

    # Проверка наличия дубликатов (базовый случай)
    if len(set(s)) == len(s): return s

    # Описанный случай замены дубликатов
    if 'zz' in s:
        z_index = s.find('zz')              # Находим индекс начала
        s = s[:z_index] + s[z_index + 2:]   # Делаем срез до и после
        s += 'a'                            # Вставляем в конец
    
    # Алгоритм замены дубликатов
    else:

        count_letter = {}           # Словарь для подсчета количества букв

        # Подсчет количества каждой буквы в строке
        for letter in alphabet:
            if letter in s:
                count_letter[letter] = s.count(letter)

        # Сортировка словаря по убыванию количества букв
        count_letter_sorted = dict(sorted(count_letter.items(), key=lambda x: x[1], reverse=True))

        # Перебор отсортированного словаря
        for letter_key, count_value in count_letter_sorted.items():

            # Если буква встречается 2 и более раз,
            if count_value >= 2:
                
                # удаляем две таких буквы из строки,
                for _ in range(2):
                    letter_index = s.find(letter_key)
                    s = s[:letter_index] + s[letter_index + 1:]

                # и добавляем следующую букву в алфавите
                next_letter = alphabet.find(letter_key) + 1

                # Формируем новую строку,
                # с проверкой выхода индекса за пределы длины алфавита
                if next_letter >= 26:
                    s += 'a'        # Если индекс выходит за пределы, добавляем 'a'
                else:
                    s += alphabet[next_letter]

            #  Прерываем цикл после первой замены буквы
            break

    # Вывод текущего шага (прогресса)
    print(f'step {step}: {s}')

    # Переход к следующему шагу (стеку)
    step += 1
    return replace_duplicates(s, step)


# Примеры использования функции
print(replace_duplicates(s='zzzab'))
print(replace_duplicates(s='hhakafh'))
print(replace_duplicates(s='zzzz'))         # Пример с большим количеством одинаковых букв
print(replace_duplicates(s='abcde'))        # Пример без дубликатов
print(replace_duplicates(s='aabbcc'))
print(replace_duplicates(s='zxyyxx'))       # Пример с выходом за пределы алфавита
````
