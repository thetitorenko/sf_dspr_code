```python
# Задание 1.1

# Прообразовываем столбец Date в формат datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Находим количество объектов, проданных во второй по популярности квартал
melb_df['Date'].dt.quarter.value_counts().values[1]
```

```python
# Задание 1.2

def get_category_encoding(df:pd.DataFrame, cols_exclude, max_unique_counts=150):
    """ Функция для преобразования столбцов датафрейма в категориальный тип,
    Args:
        df (pd.DataFrame): Датафрейм, в котором нужно преобразовать столбцы.
        cols_exclude (list): Список столбцов, которые не нужно преобразовывать.
        max_unique_counts (int, optional): Максимальное количество уникальных значений 
            для преобразования в категориальный тип. По умолчанию 150.
    Returns:
        pd.DataFrame: Датафрейм с преобразованными столбцами в категориальный тип.
    """    

    # Преобразуем столбцы в категориальный тип, 
    # если количество уникальных значений меньше max_unique_counts
    for col in df.columns:
        if df[col].nunique() < max_unique_counts and col not in cols_exclude:
            df[col] = df[col].astype('category')
    
    return df


# Преобразуем столбцы в категориальный тип, исключая указанные
cols_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
melb_df = get_category_encoding(melb_df, cols_exclude)

# Считаем кол-во категориальных столбцов
(melb_df.dtypes == 'category').sum()
```

```python
# Отсортируем данные по возрастанию цены и выведем первые 5 строк
melb_df.sort_values(by='Price').head()
```

```python
# Отсортируем данные по убыванию даты (от самой новой к самой старой)
melb_df.sort_values(by='Date', ascending=False)
```

```python
# Отсортируем данные по двум признакам (по возрастанию расстояния и цены)
# и выведем каждую 10 строку 
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

```python
# Маски для фильтрации данных
mask_type = melb_df['Type'] == 'townhouse'
mask_seller = melb_df['SellerG'] == 'McGrath'
mask_ratio = melb_df['AreaRatio'] < -0.8

# Фильтруем данные по маскам
df_filtered = melb_df[mask_type & mask_seller & mask_ratio]

# Выводим стобцы Date и AreaRatio из отфильтрованного датафрейма,
# сортируем по дате (по возрастанию) и AreaRatio (по убыванию), обновляем индексы
df_filtered.sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True).loc[:, ['Date', 'AreaRatio']]
```

```python
# Задание 2.2

# Сортируем данные по столбцу AreaRatio в порядке убывания и обновляем индексы
df_sorted = melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index=True)

# Выводим площадь здания для объекта с индексом 1558
round(df_sorted.loc[1558, 'BuildingArea'])
```

```python
# Задание 2.3

# Маски для фильтрации данных
mask_type = melb_df['Type'] == 'townhouse'
mask_rooms = melb_df['Rooms'] > 2

# Фильтруем данные по маскам
df_filtered = melb_df[mask_type & mask_rooms]

# Сортируем данные по столбцам Rooms (по возрастанию)
# и MeanRoomsSquare (по убыванию), обновляем индексы
df_sorted = df_filtered.sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True)

# Выводим стоимость объекта под индексом 18
round(df_sorted.loc[18, 'Price'])
```
