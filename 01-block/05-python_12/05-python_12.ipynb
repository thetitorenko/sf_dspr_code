```python
# Задание 1.1

# Прообразовываем столбец Date в формат datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Находим количество объектов, проданных во второй по популярности квартал
melb_df['Date'].dt.quarter.value_counts().values[1]
```

```python
# Задание 1.2

def get_category_encoding(df:pd.DataFrame, cols_exclude, max_unique_counts=150):
    """ Функция для преобразования столбцов датафрейма в категориальный тип,
    Args:
        df (pd.DataFrame): Датафрейм, в котором нужно преобразовать столбцы.
        cols_exclude (list): Список столбцов, которые не нужно преобразовывать.
        max_unique_counts (int, optional): Максимальное количество уникальных значений 
            для преобразования в категориальный тип. По умолчанию 150.
    Returns:
        pd.DataFrame: Датафрейм с преобразованными столбцами в категориальный тип.
    """    

    # Преобразуем столбцы в категориальный тип, 
    # если количество уникальных значений меньше max_unique_counts
    for col in df.columns:
        if df[col].nunique() < max_unique_counts and col not in cols_exclude:
            df[col] = df[col].astype('category')
    
    return df


# Преобразуем столбцы в категориальный тип, исключая указанные
cols_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
melb_df = get_category_encoding(melb_df, cols_exclude)

# Считаем кол-во категориальных столбцов
(melb_df.dtypes == 'category').sum()
```

### 2 юнит

```python
# Отсортируем данные по возрастанию цены и выведем первые 5 строк
melb_df.sort_values(by='Price').head()
```

```python
# Отсортируем данные по убыванию даты (от самой новой к самой старой)
melb_df.sort_values(by='Date', ascending=False)
```

```python
# Отсортируем данные по двум признакам (по возрастанию расстояния и цены)
# и выведем каждую 10 строку 
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

```python
# Маски для фильтрации данных
mask_type = melb_df['Type'] == 'townhouse'
mask_seller = melb_df['SellerG'] == 'McGrath'
mask_ratio = melb_df['AreaRatio'] < -0.8

# Фильтруем данные по маскам
df_filtered = melb_df[mask_type & mask_seller & mask_ratio]

# Выводим стобцы Date и AreaRatio из отфильтрованного датафрейма,
# сортируем по дате (по возрастанию) и AreaRatio (по убыванию), обновляем индексы
df_filtered.sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True).loc[:, ['Date', 'AreaRatio']]
```

```python
# Задание 2.2

# Сортируем данные по столбцу AreaRatio в порядке убывания и обновляем индексы
df_sorted = melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index=True)

# Выводим площадь здания для объекта с индексом 1558
round(df_sorted.loc[1558, 'BuildingArea'])
```

```python
# Задание 2.3

# Маски для фильтрации данных
mask_type = melb_df['Type'] == 'townhouse'
mask_rooms = melb_df['Rooms'] > 2

# Фильтруем данные по маскам
df_filtered = melb_df[mask_type & mask_rooms]

# Сортируем данные по столбцам Rooms (по возрастанию)
# и MeanRoomsSquare (по убыванию), обновляем индексы
df_sorted = df_filtered.sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True)

# Выводим стоимость объекта под индексом 18
round(df_sorted.loc[18, 'Price'])
```

### 3 юнит

```python
# Сделаем группировку по столбцу Type и
# посчитаем средние значения для всех числовых столбцов
melb_df.groupby(by='Type').mean(numeric_only=True)
```

```python
# Применим агрегирующую функцию и выведем среднюю цену
# для каждого типа недвижимости
melb_df.groupby(by='Type')['Price'].mean()
```

```python
# Найдем минимальное значение расстояния до центра города для каждого региона
# и отсортируем по убыванию
melb_df.groupby(by='Regionname')['Distance'].min().sort_values(ascending=False)
```

```python
# Найдем кол-во продаж, среднее, и макс. цены объектов сгруппированных по месяцу продажи
# и отсортируем по убыванию кол-ва продаж
melb_df.groupby(by='MonthSale')['Price'].agg(
    ['count', 'mean', 'max']).sort_values(by='count', ascending=False)
```

```python
# Полная статистика по цене для каждого месяца продажи
melb_df.groupby(by='MonthSale')['Price'].agg('describe')
```

```python
# Найдем количество уникальных продавцов для каждого региона,
# а также выведем список продавцов в виде множеств
melb_df.groupby(by='Regionname')['SellerG'].agg(['nunique', set])
```

```python
# Задание 3.1

# Найдем количество комнат, при котором средняя цена объектов наибольшая
melb_df.groupby(by='Rooms')['Price'].mean().sort_values(ascending=False).index[0]
```

```python
# Задание 3.2

# Найдем регион с наименьшим стандартным отклонением широты
melb_df.groupby(by='Regionname')['Lattitude'].std().sort_values().index[0]
```

```python
# Задание 3.3

# Фильтруем данные по дате продажи
start_date = pd.to_datetime('1 may 2017')
end_date = pd.to_datetime('2 sep 2017')
date_mask = melb_df['Date'].between(start_date, end_date)
melb_filtered = melb_df[date_mask]

# Найдем какая компания имеет наименьшую общую выручку за период
melb_filtered.groupby(by='SellerG')['Price'].sum().sort_values().index[0]
```

### 4 юнит

```python
# Сводная таблица показывающая зависимость медианной цены и площади от количества комнат
melb_df.groupby(by='Rooms')[['Price', 'BuildingArea']].median()
```

```python
# Сводная таблица показывающая зависимость средней цены от кол-ва комнат и типа задания
melb_df.groupby(by=['Rooms', 'Type'])['Price'].mean()
```

```python
# Отделяем вложенный уровень методом unstack()
melb_df.groupby(by=['Rooms', 'Type'])['Price'].mean().unstack()
```

```python
# Построим туже самую сводную таблицу, но с помощью метода pivot_table
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0).round()
```

```python
# Сводная таблица кол-ва продаж по регионам в будние и выходные дни
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count')
```

```python
# Найдем сред. и медианную площадь участка в зависимости от
# типа постройки и региона
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['mean', 'median'],
    fill_value=0)
```

```python
# Многомерная сводная таблица, показывающая зависимость медианной цены
# от метода продажи, типа недвижимости и региона
melb_df.pivot_table(
    values='Price',
    index=['Method', 'Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0)
```

```python
# Выводим столбцы сводной таблицы
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0)

pivot.columns
```

```python
# Получим средние значения площадей участков для типа здания unit
pivot['mean']['unit']
```

```python
# Фильтрация сводной таблицы по условию
mask = pivot['mean']['house'] < pivot['median']['house']
pivot[mask]
```
