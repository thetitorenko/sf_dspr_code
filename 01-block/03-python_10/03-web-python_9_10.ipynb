Задание 1

```python
lann_df = got_df[got_df['Allegiances'] == 'Lannister']
stark_df = got_df[got_df['Allegiances'] == 'Stark']

print(f'Персонажи верные дому Ланистеров умерли на {
    round((lann_df.shape[0] - stark_df.shape[0]) / lann_df.shape[0], 3)} раз чаще, чем персонажи верные Старкам')
```

Задание 2

```python
print('Количество персонажей умерших в цикле:')
print(f'В книге "Игра престолов": {got_df['GoT'].sum()}')
print(f'В книге "Битва королей": {got_df['CoK'].sum()}')
print(f'В книге "Буря мечей": {got_df['SoS'].sum()}')
print(f'В книге "Пир стервятников": {got_df['FfC'].sum()}')
print(f'В книге "Танец с драконами": {got_df['DwD'].sum()}')
```

Задание 3

```python
gender_list = got_df['Gender'].value_counts(normalize=True)
print(f'Соотношение погибших женщин к мужчинам: {round(gender_list[0] / gender_list[1], 2)}')
```
