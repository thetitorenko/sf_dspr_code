```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Подгрузка данных
url = 'https://raw.githubusercontent.com/thetitorenko/sf_dspr_code/main/01-block/07-python_14/data/sber_data.csv'
sber_data = pd.read_csv(url)
sber_data.head()
```

```python
# Задание 2.1

# Число строк в наборе данных
sber_data.shape[0]
```

```python
# Задание 2.2

# Сколько районов Москвы и Московской области представлено в наборе данных
sber_data['sub_area'].nunique()
```

```python
# Задание 2.3

# Максимальная цена объекта недвижимости в наборе данных
sber_data['price_doc'].max()
```

```python
# Задание 2.4

# Построение боксплота зависимости цены от экологии района
fig = plt.figure(figsize=(8, 4))
box = sns.boxplot(data=sber_data, x='price_doc', y='ecology')
box.set_title('Зависимость цены от экологии района')
fig.show()
```

```python
# Задание 2.5

fig = plt.figure(figsize=(6, 5))
scatter = sns.scatterplot(data=sber_data, x='price_doc', y='kremlin_km')
scatter.set_title('Зависимость цены от расстояния до центра города')
fig.show()
```

### 3 юнит

```python
# Булевая маска для пропущенных значений
sber_data.isnull().tail()
```

```python
# Вывод столбцов с пропущенными значениями и процентом пропусков
cols_null_percent = sber_data.isnull().mean() * 100
cols_with_null = cols_null_percent[cols_null_percent > 0].sort_values(ascending=False)
display(cols_with_null)
```

```python
# Построим столбчатую диаграмму пропусков для наглядности
cols_with_null.plot(
    kind='bar',
    figsize=(10, 4),
    title='Распределение пропусков в данных');
```

```python
# Строим тепловую карту пропусков
colors = ['blue', 'yellow']     # создаем цветовую палитру для тепловой карты
fig = plt.figure(figsize=(10, 4))
cols = cols_with_null.index     # получаем имена столбцов с пропусками
ax = sns.heatmap(
    data=sber_data[cols].isnull(),  # фильтруем только столбцы с пропусками
    cmap=sns.color_palette(colors))
ax.set_title('Тепловая карта пропусков в данных');
```
