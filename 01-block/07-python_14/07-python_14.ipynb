```python
def outliers_iqr(
        data: pd.DataFrame,
        feature: str,
        threshold: float = 1.5) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Выявление и удаление выбросов методом межквартильного размаха (IQR)

    Args:
        data (pd.DataFrame): Исходный DataFrame.
        feature (str): Имя столбца для анализа.
        threshold (float, опционально): Множитель IQR (обычно 1.5). По умолчанию 1.5

    Returns:
        tuple: (outliers, cleaned) - кортеж из двух DataFrame.
    """
    x = data[feature]

    # Расчет квартилей и IQR
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75)
    iqr = quartile_3 - quartile_1

    # Расчет границ
    lower_bound = quartile_1 - (iqr*threshold)
    upper_bound = quartile_3 + (iqr*threshold)

    # Определяем выбросы и чищенный набор данных
    outliers_mask = (x < lower_bound) | (x > upper_bound)
    outliers = data[outliers_mask].copy()
    cleaned = data[~outliers_mask].copy()

    return outliers, cleaned
```

```python
# Проверим работу функции поиска выбросов методом IQR на признаке полной площади
outliers, cleaned_sber_data = outliers_iqr(sber_data, 'full_sq')
print(f'Число выбросов по методу Тьюки: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned_sber_data.shape[0]}')
```

```python
def outliers_z_score(
        data: pd.DataFrame,
        feature: str,
        log_scale: bool = False,
        low_threshold: float = 3, 
        high_threshold: float = 3) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Выявление и удаление выбросов методом Z-оценки (метод 3-х сигм).

    Args:
        data (pd.DataFrame): Исходный DataFrame.
        feature (str): Имя столбца для анализа.
        log_scale (bool): Применять ли логарифмирование для нормализации распределения.
        low_threshold (float): Порог нижней границы (количество стандартных отклонений).
        high_threshold (float): Порог верхней границы (количество стандартных отклонений).

    Returns:
        tuple: (outliers, cleaned) - кортеж из двух DataFrame.
    """
    # Применение логарифмического преобразования при необходимости
    if log_scale:
        x = np.log1p(data[feature])
    else:
        x = data[feature]

    # Расчет среднего и стандартного отклонения
    mu, sigma = x.mean(), x.std() 

    # Расчет границ
    lower_bound = mu - (low_threshold*sigma)
    upper_bound = mu + (high_threshold*sigma)

    # Определяем выбросы и чищенный набор данных
    outliers_mask = (x < lower_bound) | (x > upper_bound)
    outliers = data[outliers_mask].copy()
    cleaned = data[~outliers_mask].copy()

    return outliers, cleaned
```

```python
# Задание 6.7

def outliers_iqr(
        data: pd.DataFrame,
        feature: str,
        log_scale: bool = False,
        low_threshold: float = 1.5, 
        high_threshold: float = 1.5) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Выявление и удаление выбросов методом межквартильного размаха (IQR)

    Args:
        data (pd.DataFrame): Исходный DataFrame.
        feature (str): Имя столбца для анализа.
        log_scale (bool): Применять ли логарифмирование (полезно для асимметричных распределений).
        low_threshold (float): Коэффициент для нижней границы.
        high_threshold (float): Коэффициент для верхней границы.

    Returns:
        tuple: (outliers, cleaned) - кортеж из двух DataFrame.
    """
    # Логарифмирование для лог-нормальных распределений
    if log_scale:
        x = np.log1p(data[feature])
    else:
        x = data[feature]

    # Расчет квартилей и IQR
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75)
    iqr = quartile_3 - quartile_1

    # Расчет границ
    lower_bound = quartile_1 - (iqr*low_threshold)
    upper_bound = quartile_3 + (iqr*high_threshold)

    # Определяем выбросы и чищенный набор данных
    outliers_mask = (x < lower_bound) | (x > upper_bound)
    outliers = data[outliers_mask].copy()
    cleaned = data[~outliers_mask].copy()

    return outliers, cleaned


outliers, cleaned_sber_data = outliers_iqr(sber_data, 'price_doc', log_scale=True, low_threshold=3, high_threshold=3)
print(f'Число выбросов по методу Тьюки: {outliers.shape[0]}')
```

```python
# Создадим алгоритм поиска неинформативных признаков

low_information_cols = []   # список для хранения

# Цикл по всем столбцам набора данных
for col in sber_data.columns:

    # Вычисляем наибольшую относительную частоту в признаке
    top_freq = sber_data[col].value_counts(normalize=True).max()

    # Вычисляем долю уникальных значений от размера признака
    nunique_ratio = sber_data[col].nunique() / sber_data[col].count()

    # сравниваем наибольшую частоту с порогом
    if top_freq > 0.95:
        low_information_cols.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
    
    # Сравниваем долю уникальных значений с порогом
    if nunique_ratio > 0.95:
        low_information_cols.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
```

```python
# Напишем функцию реализующую алгоритм выше

def get_low_informative_columns(data, threshold=0.95, verbose=False) -> list:
    """
    Функция анализирует каждый столбец и помечает его как неинформативный, 
    если он превышает заданный порог по частоте одного значения или 
    по доле уникальных значений.

    Args:
        data (pd.DataFrame): Входной датафрейм Pandas.
        threshold (float, optional): Порог отсечения признаков. По умолчанию 0.95.
        verbose (bool, optional): Флаг для вывода подробностей в консоль. 

    Returns:
        list: Список названий столбцов, определенных как неинформативные.
    """
    low_information_cols = [] 

    if verbose:
        print(f'Поиск признаков с порогом {threshold}...')

    for col in data.columns:
        # Вычисляем наибольшую относительную частоту
        top_freq = data[col].value_counts(normalize=True).max()
        
        # Вычисляем долю уникальных значений
        nunique_ratio = data[col].nunique() / data[col].count()
        
        # 1. Проверка на доминирующее значение (почти константа)
        if top_freq > threshold:
            low_information_cols.append(col)
            if verbose:
                print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
            continue 
        
        # 2. Проверка на уникальность (слишком много разных значений)
        if nunique_ratio > threshold:
            low_information_cols.append(col)
            if verbose:
                print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
                
    return low_information_cols


low_information_cols = get_low_informative_columns(sber_data, verbose=True)
```

```python
# Удаляем неинформативные признаки из набора данных
information_sber_data = sber_data.drop(axis=1, labels=low_information_cols)
print(f'Результирующее число признаков: {information_sber_data.shape[1]}')
```
